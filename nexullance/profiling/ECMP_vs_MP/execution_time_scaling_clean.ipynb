```python
import pandas as pd
# import matplotlib.pyplot as plt
import numpy as np
import os
import sys
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from plotting_tempelate import plot_scaling_lines
import ast
import matplotlib.pyplot as plt
```

```python
# Load data files into a big dictioanry
raw_data = {}
traffics = []
for file in os.listdir("."):
    if file.endswith('.csv'):
        name = (file.split('.')[0]).split('_')[-1]
        # print(name)
        raw_data[name] = pd.read_csv(os.path.join(file))
        traffics.append(name)
```

```python
traffic = "uniform"
_V = list(raw_data[traffic]["V"])
methods = list(raw_data[traffic].keys())[2:]

plot_data={method: [] for method in methods}
for method in plot_data.keys():
    plot_data[method] = raw_data[traffic][method]

_x_power = 1
plot_scaling_lines(_V, plot_data, "scaling of routing methods with "+traffic + " traffic" + "\n in jellyfish networks", "V^"+ r"$\bf{" + str(_x_power) + "}$", "execution time (s)", True, _x_power)
plt.show()
_x_power = 5
plot_scaling_lines(_V, plot_data, "scaling of routing methods with "+traffic + " traffic" + "\n in jellyfish networks", "V^"+ r"$\bf{" + str(_x_power) + "}$", "execution time (s)", False, _x_power)
plt.show()

plot_data={method: [] for method in methods if method not in ['Nexullance_MP_APST_4']}
for method in plot_data.keys():
    plot_data[method] = raw_data[traffic][method]
_x_power = 3
plot_scaling_lines(_V, plot_data, "scaling of routing methods with "+traffic + " traffic" + "\n in jellyfish networks", "V^"+ r"$\bf{" + str(_x_power) + "}$", "execution time (s)", False, _x_power)
plt.show()

plot_data={method: [] for method in methods if method not in ['Nexullance_MP_APST_4', 'ECMP_APST_4']}
for method in plot_data.keys():
    plot_data[method] = raw_data[traffic][method]
_x_power = 3
plot_scaling_lines(_V, plot_data, "scaling of routing methods with "+traffic + " traffic" + "\n in jellyfish networks", "V^"+ r"$\bf{" + str(_x_power) + "}$", "execution time (s)", False, _x_power)
plt.show()
```

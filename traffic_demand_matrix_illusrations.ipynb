```python
import globals as gl
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 20})
```

```python
V = 8
D = 3
EPR = (D+1)//2
```

```python
# show the 2D array as a heatmap
plt.imshow(gl.generate_uniform_traffic_pattern(V, EPR))
# plt.colorbar(label='relative flow intensity')
plt.xlabel('source EP id')
plt.ylabel('destination EP id')
plt.title('uniform')
plt.show()
```

```python
# show the 2D array as a heatmap
plt.imshow(gl.generate_diagonal_traffic_pattern(V, EPR, 1))
# plt.colorbar(label='relative flow intensity')
plt.xlabel('source EP id')
plt.ylabel('destination EP id')
plt.title('nearst-neighbour')
plt.show()
```

```python
# show the 2D array as a heatmap
plt.imshow(gl.generate_shift_traffic_pattern(V, EPR, 1))
plt.colorbar(label='relative flow intensity')
plt.xlabel('source EP id')
plt.ylabel('destination EP id')
plt.title('shift-1')
plt.show()
```

```python
# show the 2D array as a heatmap
plt.imshow(gl.generate_half_shift_traffic_pattern(V, EPR))
# plt.colorbar(label='relative flow intensity')
plt.xlabel('source EP id')
plt.ylabel('destination EP id')
plt.title('shift-half')
plt.show()
```

```python
# show the 2D array as a heatmap
plt.imshow(gl.generate_uniform_cluster_pattern(V, EPR, 4))
# plt.colorbar(label='relative flow intensity')
plt.xlabel('source EP id')
plt.ylabel('destination EP id')
plt.title('router-clustered')
plt.show()
```

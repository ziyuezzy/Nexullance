```python
import globals as gl
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})
```

```python
TM=gl.generate_uniform_traffic_pattern(8,2)
# illustrate traffic pattern:
plt.matshow(TM, cmap='viridis')
plt.colorbar(label='relative flow intensity')
plt.title('uniform $TM^R$')
plt.xlabel('source routers')
plt.ylabel('destination routers')
plt.show()
```

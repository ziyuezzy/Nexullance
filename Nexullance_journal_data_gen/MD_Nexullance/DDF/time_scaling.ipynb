```python
MP_times = df["time_MP[s]"]
MP_tot_time = sum(MP_times)
MP_peak_RAM = np.average(df["peak_RAM_MP[MB]"])

IT_times = df["time_IT[s]"]
IT_tot_time = sum(IT_times)
IT_peak_RAM = np.average(df["peak_RAM_IT[MB]"])
```

```python
data_points = [MP_tot_time, MD_MP_time, IT_tot_time, MD_IT_time]
labels = ['Nexullance_MP_APST4 (SUM)', 'MD-Nexullance_MP_APST4', 'Nexullance_IT (SUM)', 'MD-Nexullance_IT']

# Calculate bar width dynamically based on the number of methods
# Generating x positions for each group of bars
x_positions = np.arange(len(data_points))  


plt.bar(x_positions, data_points, align='center', label=labels, color=['r', 'b', 'g', 'y'])
plt.yscale("log")  
plt.xticks(x_positions, [])
plt.ylabel('Calculation Time [s]')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
# plt.show()
plt.title("Caculation Time for four routing methods, \n with the shift family of traffic demand matrices")
```

```python
from globals import *
import matplotlib.pyplot as plt
import topologies.RRG as RRG
import topologies.Slimfly as Slimfly
# import topologies.Equality as Equality
# import topologies.GDBG as GDBG
# import topologies.fake_GDBG as fake_GDBG
import pickle
import math
```

```python
config=(98,11)
topo_name='slimfly'
path_dict = pickle.load(open(f'./pickled_data/graphs_and_paths/all_shortest_paths_{config}_{topo_name}_paths.pickle', 'rb'))
edgelist = pickle.load(open(f'./pickled_data/graphs_and_paths/{config}_{topo_name}_edgelist.pickle', 'rb'))
_network=Slimfly.Slimflytopo(edgelist)

_x = []
_min_link_load = []
_ave_link_load = []
_max_link_load = []
_min_local_link_load = []
_ave_local_link_load = []
_max_local_link_load = []
for subcription in range(1, config[1]):
    _x.append(subcription)
    link_load, local_link_load = _network.distribute_uniform_flow_on_paths_with_EP(path_dict, subcription)
    

    _max_local_link_load.append(max(list(local_link_load))/mean(list(local_link_load)))
    _ave_local_link_load.append(mean(list(local_link_load))/mean(list(local_link_load)))
    _min_local_link_load.append(min(list(local_link_load))/mean(list(local_link_load)))

    _max_link_load.append(max(list(link_load))/mean(list(local_link_load)))
    _ave_link_load.append(mean(list(link_load))/mean(list(local_link_load)))
    _min_link_load.append(min(list(link_load))/mean(list(local_link_load)))
    
```

```python
# Create the scatter plot
plt.scatter(_x, _ave_link_load, color='red', label='R2R link load')
for i in range(len(_x)):
    plt.errorbar(_x[i], _ave_link_load[i], yerr=[[_ave_link_load[i]-_min_link_load[i]], 
                            [_max_link_load[i]-_ave_link_load[i]]], fmt='none', ecolor='red', capsize=4)
plt.scatter(_x, _ave_local_link_load, color='blue', label='local link load')
for i in range(len(_x)):
    plt.errorbar(_x[i], _ave_local_link_load[i], yerr=[[_ave_local_link_load[i]-_min_local_link_load[i]], 
                            [_max_local_link_load[i]-_ave_local_link_load[i]]], fmt='none', ecolor='blue', capsize=4)

# optimal subcription in the slimfly paper:
p_optimal=math.ceil(config[1]/2)
plt.axvline(x = p_optimal, color = 'g', label = 'theoratical optimal of load balancing (slimfly2016)', linestyle='dashed')


# Add labels and title
plt.xlabel('router subscription')
plt.ylabel('link load')
plt.title(f'local and R2R link load of {config}_{topo_name}')
plt.legend(loc='upper left')


# Show the plot
plt.show()
```

```python
from globals import *
import matplotlib.pyplot as ax
import topologies.RRG as RRG
import topologies.Slimfly as Slimfly
import pickle
import math
import numpy as np
```

```python
config=(98,11)
topo_name='jellyfish'
path_dict = pickle.load(open(f'./pickled_data/graphs_and_paths/all_shortest_paths_{config}_{topo_name}_paths.pickle', 'rb'))
edgelist = pickle.load(open(f'./pickled_data/graphs_and_paths/{config}_{topo_name}_edgelist.pickle', 'rb'))
_network=Slimfly.Slimflytopo(edgelist)

_x = []
_min_link_load = []
_ave_link_load = []
_max_link_load = []
local_link_load = 0
for subcription in range(1, config[1]):
    _x.append(subcription)
    link_load, local_link_load = _network.distribute_uniform_flow_on_paths_with_EP(path_dict, subcription)
    
    _max_link_load.append(max(list(link_load))/local_link_load)
    _ave_link_load.append(mean(list(link_load))/local_link_load)
    _min_link_load.append(min(list(link_load))/local_link_load)
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import math
from IPython.core.display import display, HTML

# Create a function to update the plot for each frame
def update_frame(offered_load):
    ax.clear()  # Clear the current figure
    ax.scatter(_x, offered_load * np.array(_ave_link_load), color='red', label='R2R link load')
    for i in range(len(_x)):
        ax.errorbar(_x[i], offered_load * _ave_link_load[i], yerr=[[offered_load * _ave_link_load[i] - offered_load * _min_link_load[i]],
                                                                    [offered_load * _max_link_load[i] - offered_load * _ave_link_load[i]]],
                     fmt='none', ecolor='red', capsize=4)
    
    p_optimal = math.ceil(config[1] / 2)
    # ax.axvline(x=p_optimal, color='g', label='theoretical optimal of load balancing (slimfly2016)', linestyle='dashed')
    ax.axhline(y=1.0, color='b', label='local link capacity = R2R link capacity', linestyle='solid')
    ax.axhline(y=offered_load, color='b', label='offered load', linestyle='dashed')
    ax.set_ylim([0, 2.0])
    ax.legend(loc='upper left')


# Initialize the plot
fig, ax = plt.subplots()
ax.set_xlabel('router subscription')
ax.set_ylabel('link load')
# ax.set_ylim([0, 2.0])
ax.set_title(f'local and R2R link load of {config}_{topo_name}')

ani = animation.FuncAnimation(fig, update_frame, frames=np.arange(0, 1.1, 0.1))

plt.close()
HTML(ani.to_jshtml())
```

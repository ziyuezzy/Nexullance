```python
import topologies.RRG as RRG
import globals as gl
import networkx as nx
import matplotlib.pyplot as plt
```

```python
config=(8, 4)
EPR=3
topo='jellyfish'
_network=RRG.RRGtopo(config[0], config[1])
```

```python
G=_network.nx_graph

# Generate labels for vertices
labels = {node: f"R{node+1}" for node in G.nodes()}

# Draw the graph
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, labels=labels, node_size=1500, node_color='lightblue', font_size=20, node_shape='s')
plt.title("jellyfish")
plt.show()
```
